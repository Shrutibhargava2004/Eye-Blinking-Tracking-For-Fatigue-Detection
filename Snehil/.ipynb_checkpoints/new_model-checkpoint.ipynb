### Labeling

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
DataDirectory_train="D:\\MINOR\\Project\\my_dataset\\train\\"
Classes=["Active","Sleepy"]
for category in Classes:
    path=os.path.join(DataDirectory_train,category)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img))
        plt.imshow(img_array)
        plt.show()
        break
    break
```

### Reading all images and converting them into array for data and labels

```python
training_Data=[]

def create_training_Data():
    with open("D:\\MINOR\\Project\\my_dataset\\train\\annotations.txt", "w") as f:
        
        for category in Classes:
            path=os.path.join(DataDirectory_train,category)
            class_num=Classes.index(category)
            for img in os.listdir(path):
                try:
                    img_array=cv2.imread(os.path.join(path,img))
                    new_array=cv2.resize(img_array, (224,224))
                    training_Data.append([new_array,class_num])
                    annotation=f"{img}\t{class_num}\n"
                    f.write(annotation)
                except Exception as e:
                    pass
```

```python
create_training_Data()
```

```python
x_train=[]
y_train=[]

for features, label in training_Data:
    x_train.append(features)
    y_train.append(label)
X_train=np.array(x_train)
Y_train=np.array(y_train)
```

```python
X_train.shape, Y_train.shape
```

```python
import pickle
pickle_out=open("X_train.pickle","wb")
pickle.dump(X_train, pickle_out)
pickle_out.close()

pickle_out=open("Y_train.pickle","wb")
pickle.dump(Y_train, pickle_out)
pickle_out.close()
```

```python
pickle_in=open("X_train.pickle","rb")
X_train=pickle.load(pickle_in)

pickle_in=open("Y_train.pickle","rb")
Y_train=pickle.load(pickle_in)
```

```python
DataDirectory_valid="D:\\MINOR\\Project\\my_dataset\\valid\\"
valid_Data=[]

def create_valid_Data():
    with open("D:\\MINOR\\Project\\my_dataset\\valid\\annotations.txt", "w") as f:
        
        for category in Classes:
            path=os.path.join(DataDirectory_valid,category)
            class_num=Classes.index(category)
            for img in os.listdir(path):
                try:
                    img_array=cv2.imread(os.path.join(path,img))
                    new_array=cv2.resize(img_array, (224,224))
                    valid_Data.append([new_array,class_num])
                    annotation=f"{img}\t{class_num}\n"
                    f.write(annotation)
                except Exception as e:
                    pass
create_valid_Data()
```

### Deep learning model for training

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model=tf.keras.applications.mobilenet.MobileNet()
```

```python
model.summary()
```

### Transfer learning

```python
base_input=model.layers[0].input
```

```python
base_output=model.layers[-4].output
```

```python
Flat_layer=layers.Flatten()(base_output)
final_output=layers.Dense(1)(Flat_layer)
final_output=layers.Activation('sigmoid')(final_output)
```

```python
new_model=keras.Model(inputs=base_input, outputs=final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
history=new_model.fit(X_train,Y_train,validation_data=(X_valid,Y_valid),epochs=4)
```

```python
# Plot loss
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.title('Loss vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Plot accuracy
plt.plot(history.history['accuracy'], label='Training accuracy')
plt.plot(history.history['val_accuracy'], label='Validation accuracy')
plt.title('Accuracy vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
def calculate_metrics(ground_truth, predicted_mask):
    # Flatten the binary masks to 1D arrays
    ground_truth = ground_truth.ravel()
    predicted_mask = predicted_mask.ravel()

    # True Positives, False Positives, False Negatives
    TP = np.sum((ground_truth == 1) & (predicted_mask == 1))
    FP = np.sum((ground_truth == 0) & (predicted_mask == 1))
    FN = np.sum((ground_truth == 1) & (predicted_mask == 0))

    # True Negatives (may not be needed)
    TN = np.sum((ground_truth == 0) & (predicted_mask == 0))

    # Calculate Precision, Recall, and F1 Score
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    f1_score = 2 * (precision * recall) / (precision + recall)

    # Calculate Intersection over Union (IoU)
    iou = TP / (TP + FP + FN)

    return precision, recall, f1_score, iou
```

```python
calculate_metrics(Y_test,prediction)
```
